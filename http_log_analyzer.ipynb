```python
import csv
import pandas as pd
import io
from datetime import datetime,timedelta
```

### Assumtion: a report is generated for a particular period of time by the requester 

##### Going line by line avoids loading the enter file into memory
###### Because time component is set to 00:00:00, adding a day and subtracting a second includes the entire end date

```python
def web_file_analyzer(file,start_date_str,end_date_str):
    date_format = '%m/%d/%Y'
    start_date = datetime.strptime(start_date_str,date_format)
    end_date = datetime.strptime(end_date_str,date_format)
    end_date = end_date + timedelta(days=1) - timedelta(seconds=1)
    df_pages_unique = pd.DataFrame(columns=('Date','Page','Hits'))
    df_users_unique = pd.DataFrame(columns=('Date','User','Hits'))
    total_pages = 0
    with open(file) as logfile:
        for line in logfile:
            cols = line.split(',')
            line_date = datetime.strptime(cols[2][:10],'%Y-%m-%d')
            if start_date <= line_date <= end_date:
                df_pages_unique.loc[total_pages] = [cols[2][:10],cols[0], 1]
                df_users_unique.loc[total_pages] = [cols[2][:10],cols[1], 1]
                total_pages +=1
        return (df_pages_unique,df_users_unique)
```

##### For our porposes, we'll pass the function values as if passed by a UI mechanism, which returns a tuple of Pandas DataFrames

```python
df_pages, df_users = web_file_analyzer("http_data_sample.csv","01/01/2018","01/04/2018")
```
